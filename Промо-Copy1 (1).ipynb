```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sales=pd.read_csv("C:\Работа\Анализ продаж.csv",sep=",",low_memory=False)
promo=pd.read_csv('C:\\Работа\\excelReport.csv',sep=';')

#sales=sales.dropna(axis='columns')#Удалили ненужные столбцы

#display(sales.head())
#display(sales.head(5))
#display(promo.head(1))

#sales.info()
#promo.info()
```

## Предобработка данных

```python
sales.columns=[i.lower() for i in sales.columns]

sales["dates"]=pd.to_datetime(sales['dates'],format='%d.%m.%Y')
sales['day']=pd.DatetimeIndex(sales['dates']).day
sales['sls_f_sale_amt_novat']=sales['sls_f_sale_amt_novat'].apply(lambda x: x.replace(',','.'))
sales['sls_f_sale_amt_novat']=pd.to_numeric(sales['sls_f_sale_amt_novat'])
promo['Скидка']=promo['Скидка'].apply(lambda x: x.replace(',','.'))
promo['Скидка']=pd.to_numeric(promo['Скидка'])
promo['Скидка']=promo['Скидка'].astype('int')
sales['margin_prc']=sales['margin_prc'].str.strip('%')
sales['margin_prc']=pd.to_numeric(sales['margin_prc'])
```

```python
def coof(promo):
    if promo['Скидка']>25:
        return 2.5
    elif promo['Скидка']==25:
        return 2
    elif promo['Скидка']==20:
        return 1.75
    elif promo['Скидка']<20:
        return 1.5
promo['Коофицент']=promo.apply(coof,axis=1) 

for i in promo['Артикул']:
    if sales.query('article_bk==@i')['f_sale_unit_qty'].mean()>12:
            coof=0.5+promo.query('Артикул==@i')['Коофицент']
            name=promo.query('Артикул==@i')['Наименование.1']
            dis=(promo.loc[promo["Артикул"]==i]["Скидка"]);
            print(f"Акртикул {i} {name}")
            print('Скидка на этот товар',dis)
            
            print(f'Общее количество за месяц {sales.query("article_bk==@i")["f_sale_unit_qty"].sum()},на сумму в {sales.query("article_bk==@i")["sls_f_sale_amt_novat"].sum()}')
            print(f'В сренем продается {np.round(sales.query("article_bk==@i")["f_sale_unit_qty"].mean(),1)} eд,на сумму {np.round(sales.query("article_bk==@i")["sls_f_sale_amt_novat"].mean(),1)}' )
            print('Рекомендуем на неделю',np.ceil((sales.query('article_bk==@i')['f_sale_unit_qty'].mean())*7*coof))
            print()
    elif sales.query('article_bk==@i')['f_sale_unit_qty'].mean()<12:
            coof=0+promo.query('Артикул==@i')['Коофицент']
            print(f"Акртикул {i} {promo.query('Артикул==@i')['Наименование.1']}")
            print('Скидка на этот товар',promo.loc[promo["Артикул"]==i]["Скидка"])
            
            print(f'Общее количество за месяц {sales.query("article_bk==@i")["f_sale_unit_qty"].sum()},на сумму в {sales.query("article_bk==@i")["sls_f_sale_amt_novat"].sum()}')
            print(f'В сренем продается {np.round(sales.query("article_bk==@i")["f_sale_unit_qty"].mean(),1)} eд,на сумму {np.round(sales.query("article_bk==@i")["sls_f_sale_amt_novat"].mean(),1)}' )
            print('Рекомендуем на неделю',np.ceil((sales.query('article_bk==@i')['f_sale_unit_qty'].mean())*7*coof))
            print()
```
